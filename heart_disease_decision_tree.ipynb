```python
#Import Dependencies 
import pandas as pd
import os
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#Load Data
heart_csv = "Resources/heart.csv"
heart_df = pd.read_csv(heart_csv)
heart_df.head(10)
```

```python
target = heart_df["sex"]
target_names = ["0", "1"]
```

```python
data = heart_df.drop("sex", axis=1)
feature_names = data.columns
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target)
```

```python
#Decision Tree training & Score 
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(f"Decision Tree Score: {clf.score(X_test, y_test)}")
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf = rf.fit(X_train, y_train)
print(f"Random Forest Score: {rf.score(X_test, y_test)}")
```
